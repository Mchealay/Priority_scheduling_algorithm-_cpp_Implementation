```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('heart.data.csv')

# Generate data for plotting
min_smoking = data['smoking'].min()
mean_smoking = data['smoking'].mean()
max_smoking = data['smoking'].max()

x = np.linspace(data['biking'].min(), data['biking'].max(), 100)
x_const_min_smoking = np.column_stack((np.full_like(x, min_smoking), x))
x_const_mean_smoking = np.column_stack((np.full_like(x, mean_smoking), x))
x_const_max_smoking = np.column_stack((np.full_like(x, max_smoking), x))

# Add a constant term for the intercept
x_const_min_smoking = sm.add_constant(x_const_min_smoking, has_constant='add')
x_const_mean_smoking = sm.add_constant(x_const_mean_smoking, has_constant='add')
x_const_max_smoking = sm.add_constant(x_const_max_smoking, has_constant='add')

# Fit the OLS model
model = sm.OLS(data['heart.disease'], sm.add_constant(data[['smoking', 'biking']])).fit()

# Predict the values
y_pred_min_smoking = model.predict(x_const_min_smoking)
y_pred_mean_smoking = model.predict(x_const_mean_smoking)
y_pred_max_smoking = model.predict(x_const_max_smoking)

# Plotting
plt.figure(figsize=(12, 8))

plt.scatter(data['biking'], data['heart.disease'], color='blue', label='Observed data')
plt.plot(x, y_pred_min_smoking, color='red', label='Min Smoking')
plt.plot(x, y_pred_mean_smoking, color='green', label='Mean Smoking')
plt.plot(x, y_pred_max_smoking, color='purple', label='Max Smoking')

plt.xlabel('Percentage of People Biking to Work')
plt.ylabel('Heart Disease')
plt.title('Effect of Biking on Heart Disease with Different Smoking Rates')
plt.legend()
plt.grid(True)
plt.show()
```
